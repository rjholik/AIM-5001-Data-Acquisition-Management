### Question 4
(15 pts)

Describe any modifications to existing Chinook tables that you will be making to enable the downloading of movie content by Chinook customers. Create and share the SQL scripts required for those modifications. 

### Solution:

We will add a new column "MovieId" to the "InvoiceLine" table. Then we change the "NOT NULL" constraint for the "TrackID" column. With this, we can start adding new records to this table with a link to the "Movie" table and with "NULL" in the "TrackId" column. The necessary condition is that one of the columns "TrackId" or "MovieId" always has a record, thus preserving the integrity of the table ''Track" or "Movie".

ALTER TABLE "InvoiceLine" ADD COLUMN "MovieId" int;

ALTER TABLE "InvoiceLine" ALTER COLUMN "TrackId" DROP NOT NULL;
